# 🎭 Abstraction in Python

## 📌 Introduction
Abstraction is one of the fundamental principles of Object-Oriented Programming (OOP), along with Encapsulation, Inheritance, and Polymorphism. It is the concept of hiding complex implementation details and showing only the essential features of an object.

In simpler terms, abstraction lets you focus on **what** an object does instead of **how** it does it.

---

## 🧠 Why Use Abstraction?
- Improves code readability and maintainability
- Promotes modularity by separating interface from implementation
- Allows implementation changes without affecting external code
- Encourages code reuse and scalability

---

## 🛠 Implementing Abstraction in Python

Python achieves abstraction using:
- Abstract Base Classes (ABCs)
- The abc module

### ✅ Abstract Class
An abstract class cannot be instantiated. It may contain one or more abstract methods.

### ✅ Abstract Method
An abstract method is declared in the abstract class but doesn't have an implementation. Subclasses are required to implement it.

---

## 🔍 Syntax

```python
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract class
    @abstractmethod
    def make_sound(self):  # Abstract method
        pass

class Dog(Animal):
    def make_sound(self):
        return "Bark"

class Cat(Animal):
    def make_sound(self):
        return "Meow"
````

> 💡 Note: Attempting to instantiate an abstract class directly will result in a TypeError.

---

## 🔗 Key Points

* Use `from abc import ABC, abstractmethod`
* Inherit from `ABC` to make a class abstract
* Use `@abstractmethod` to define abstract methods
* Subclasses must implement all abstract methods, or they will also be abstract

---

## 🎯 Real-Life Example

Think of a **remote control**: You use buttons like `Volume Up` and `Power`, but you don't know the complex circuitry behind them. That’s abstraction — using functionality without knowing the underlying complexity.

---

## ✅ When to Use Abstraction?

* When creating a blueprint for a family of classes
* When defining common interfaces that must be implemented
* When working in teams to define interfaces and expectations

---

## 📚 Summary

| Concept         | Description                                 |
| --------------- | ------------------------------------------- |
| Abstract Class  | Cannot be instantiated, may contain methods |
| Abstract Method | Must be implemented in subclass             |
| abc module      | Enables abstract class creation             |

Abstraction provides a clean and scalable way to define high-level logic while hiding implementation details. It's a powerful tool for designing robust and maintainable code.

---

`````python
from abc import ABC,abstractmethod

# Abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    
car=Car()
operate_vehicle(car)  
`````

`````python
# 🧑‍💻 Create an abstract class Vehicle with an abstract method `start_engine()`.
#  Then create two subclasses (Car and Truck) that implement the method.


from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        print("normal")
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")
class Truck(Vehicle):
    def start_engine(self):
        print("Truck engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)
truck = Truck()
operate_vehicle(truck)

`````
